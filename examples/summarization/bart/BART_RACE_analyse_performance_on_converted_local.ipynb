```python
from pathlib import Path
import json
```

```python
# read input data
set_ = 'dev'
input_file = Path(f'../../../../qa-to-nli/.data/RACE/set3_neural/nli_in_qa_format_{set_}.json')
preds_file = Path(f'../../../../qa-to-nli/eval_preds.txt')
ids_to_remove_file = Path(f'./bad_hypothesis_ids.txt')
```

```python
with open(input_file) as f:
    samples = json.load(f)
```

```python
with open(preds_file) as f:
    preds = [int(v) for v in f.read().split(',')]
```

```python
print(f"length preds: {len(preds)}, num examples: {len(samples)}")
```

```python
def get_accuracy(samples, preds, filter_=None):
    if filter_ is None:
        filter_ = lambda x: True
    assert len(samples) == len(preds)
    true = 0
    total = 0
    for s,p in zip(samples, preds):
        if filter_(s):
            total+=1
            if s['label'] ==p:
                true+=1
    print(f"total={total}, true={true}")
    return true/total
        
```

```python
get_accuracy(samples, preds)
```

```python
with open(ids_to_remove_file) as f:
    idx_to_remove = set(f.read().split())
```

```python
print(f'num to remove: {len(idx_to_remove)}')
```

```python
get_accuracy(samples, preds, filter_=lambda x: x['id'] not in idx_to_remove)
```
